```python
# 이미지 클릭

def click_img_area(xpath) :

    try :
        click_img = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, xpath)))
        click_img.click()

    except Exception as e:
        print(f'에러발생 {e}')
        print(f'새로고침 실행')
        driver.refresh()
        time.sleep(2)

        return click_img_area(xpath)

    return click_img
    
```

```python
# topic_select 

def get_scroll() :
    pos = driver.execute_script("return window.pageYOffset;")
    
    return pos

def scroll_down() :
    last_pos = 0
    pause_cnt = 0
    max_pause_cnt = 20
    
    while True : 
        driver.execute_script("window.scrollBy({top : window.innerHeight * 2, behavior : 'smooth'})")
        time.sleep(0.5)

        new_pos = get_scroll()

        if last_pos == new_pos :
            pause_cnt += 1

        else :
            pause_cnt = 0
            last_pos = new_pos    

        if pause_cnt >= max_pause_cnt :
            break


# scroll_down()
```

```python
# 네이버

def naver_img_links(topic, option = '', limit = 20) :

    if naver :
        driver.get(f'https://search.naver.com/search.naver?&where=image&query={topic}&{option}')
    
    time.sleep(2)
    click_img_area('//div[@class="tile_item _fe_image_tab_content_tile"]//img[@class="_fe_image_tab_content_thumbnail_image"]')
    next_button = driver.find_element(By.TAG_NAME, 'body')
    
    links = []
    cnt = 1
    
    while len(links) < limit :
        img_xpath = '//div[@class="image"]//img'
        t1 = time.time()

        while True :
            imgs = driver.find_elements(By.XPATH, img_xpath)
            t2 = time.time()

            if len(imgs) > 0 :
                break

            if t1 - t2 > 5 :
                print(f'이미지 찾기 실패 {imgs}')
                break

        try :
            src = imgs[0].get_attribute('src')

            if src is not None and src not in links :
                links.append(src)
                print(f'{cnt}, {src}')
                
                cnt += 1
        except StaleElementReferenceException :
            print(f'{StaleElementReferenceException} 발생')
            pass
        
        except Exception as e :
            print(f'네이버 크롤링 에러 발생 {e}')

        next_button.send_keys(Keys.RIGHT)
        
    print('=' * 25 + '네이버 크롤링 완료' + '=' * 25)
    print(f'topic : {topic} 이미지 수 : {limit}개 네이버 크롤링 완료')
    print('=' * 23 + '=' * len('네이버 크롤링 완료') * 2 + '=' * 23)
    return links
```

```python
# json 파일 

def make_json_browser() :
    file_path = os.getcwd()
    
    data = {
        "google": {
            "search_url": "https://www.google.com/search?q={topic}&tbm=isch{option}",
            "click_xpath": "//div[@jsname='dTDiAc']",
            "img_xpath": "//div[@jsname='figiqf']//img"
        },
        "naver" : {
            "search_url": "https://search.naver.com/search.naver?&where=image&query={topic}&{option}",
            "click_xpath": "//div[@class='tile_item _fe_image_tab_content_tile']//img[@class='_fe_image_tab_content_thumbnail_image']",
            "img_xpath": "//div[@class='image']//img"
        }
    }
    
    if not os.path.exists(os.path.join(file_path, 'browser.json')) :
    
        with open('browser.json', 'w', encoding = 'utf-8') as f :
            json.dump(data, f, indent = 4)
            print(f'데이터가 {file_path} 경로에 json 형식으로 올바르게 저장되되었습니다.')
            
    else :
        print('현재 경로에 파일 존재')
    
def read_json_browser():

    with open('browser.json', 'r', encoding='utf-8') as f:
        browser_data = json.load(f)
        keys = list(browser_data.keys())
    
    return browser_data, keys

def check_json_browser(browser) :
    file_path = os.getcwd()
    
    try :
        browser_data, keys = read_json_browser()
            
        if browser not in keys :

            raise KeyError(f'{browser} 브라우저 데이터가 존재하지 않음')

        check_browser = browser_data[browser] 

        for data in ['search_url', 'click_xpath', 'img_xpath'] :

            if data not in check_browser :

                raise KeyError(f'{browser} 브라우저에 대한 {data} 데이터가 존재하지 않음')
            
    except FileNotFoundError:
        print(f'올바른 경로에 browser.json 파일이 존재하지 않습니다. 파일 경로를 확인하세요.')
        
        return 
    
    except json.JSONDecodeError:
        print('파일 내용이 올바른 JSON 형식이 아닙니다. 파일을 확인하세요.')
    
        return 
    
    except KeyError as e :
        print(f'브라우저 에러 : {e}')
        
        return 
    
    except Exception as e :
        print(f'예상 외의 에러 {e} 발생')
    
        return 
    
    return browser_data[browser]

make_json_browser()
browser_data, _ = read_json_browser()
print(list(browser_data.keys()))
print(os.getcwd())
```

```python
# 혼합

def img_links(browser, topic, option = '', limit = 20) :
    search_data = check_json_browser(browser)
    
    driver.get(search_data['search_url'].format(topic=topic, option=option))
    time.sleep(2)
    click_img_area(search_data['click_xpath'])
    img_xpath = search_data['img_xpath']

    next_button = driver.find_element(By.TAG_NAME, 'body')
    
    links = []
    prev_len = 0
    prev_len_cnt = 10
    cnt = 0
    limit = 100 if limit == 0 else limit
    
    while len(links) < limit :
        t1 = time.time()    

        while True :
            imgs = driver.find_elements(By.XPATH, img_xpath)
            t2 = time.time()

            if len(imgs) > 0 :
                break

            if t1 - t2 > 5 :
                print(f'이미지 찾기 실패 {imgs}')
                break
        
        if not imgs :
            print(f'이미지 찾기 실패 {imgs}')

            continue
        
        retry_cnt = 3
        
        while retry_cnt > 0 :
        
            try :
                src = imgs[0].get_attribute('src')

                if src is not None and src not in links :
                    links.append(src)
                    cnt += 1
                    
                    print(f'{cnt}, {src}')
                    
                if len(links) == prev_len:
                    prev_len_cnt -= 1

                else :
                    prev_len_cnt = 10

                prev_len = len(links)                    
                    
                break

            except StaleElementReferenceException :
                print(f'{StaleElementReferenceException} 발생')
                print(src)
                imgs = driver.find_elements(By.XPATH, img_xpath)
                retry_cnt -= 1
                time.sleep(1)
            
            except Exception as e :
                print(f'네이버 크롤링 에러 발생 {e}')
                
                break
        
        if retry_cnt == 0 :
            print(f'{cnt}번째 이미지 로드 실패 다음 이미지로 넘어감')
            
        if prev_len_cnt <= 0 :
            print('재시도 끝에 새로운 이미지를 찾을 수 없어 크롤링을 종료 합니다.')

            break
            
        next_button.send_keys(Keys.RIGHT)
    
    print('=' * 100)
    print(f'{browser.capitalize()} 크롤링 완료')
    print(f'Browser : {browser.capitalize()}, Topic : {topic}, 이미지 수 : {cnt}개')
    print('=' * 100)
    
    return links
```
